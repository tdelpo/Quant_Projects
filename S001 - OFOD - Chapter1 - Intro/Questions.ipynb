**Q1.1** - Explain carefully the difference between selling a call option and buying a put option.

Answer: 
 - A call option is the right, but not obligation, to buy the underlying at an certain price (strike) by a certain date (expiry/maturity). If you sell a call option, you are selling to another party, the right to buy. As the seller, you will be obligated, if exercised, to sell the underlying to the other party at the agreed price. 
 - The payoff from selling a call option is: \
        $Payoff = -max(0, Spot(S) - Strike(K)) = min(Strike - Spot, 0)$
 - A put option is the right, but not obligation, to sell the underlying at a certain price (strike) by a certain date (expiry/maturity). If you buy a put options, you are purchasing the right to sell. The selling of the option is obligated to buy your underlying if exercised. 
 - The payoff from buying a put option is: \
        $Payoff = max(0, Strike - Spot)$

**Q1.8** - Explain why a futures contract can be used for either speculation or hedging.

Answer:
 - A futures contract allows a party to enter into an agreement with another party (directly or via exchange/clearing house) to exchance an asset at a certain date in the future for a certain price. 
 - A speculator, who believe an asset will increase in value in time, can enter into a long future contract (agreement to buy) an asset for a price below where they speculate the future price will be. Upon receiving the underlying asset, they can sell immediately for a profit. Similarly, they can agree to sell an asset in the future for a price less than they believe the spot price will be on that day.
 - A hedger can protect their downside through futures contracts. If looking to sell an asset in the future and believing the spot price may decrease in the meantime, they can enter into a futures contract for a price which insures today's current value. 

**Q1.9** - Suppose that a March call option to buy a share for $50 costs $2.50 and is held until March. Under what circumstances will the holder of the option make a profit? Under what circumstances will the option be exercised> Draw a diagram illustrating how the profit from a long position in the option depends on the stock price at maturity of the option.

Answer:
 - The holder of the option is the party that is 'long', e.g. they have bought the option to buy the shares at $50.
 - The holder of this option makes a profit when the underlying exceeds $52.50 (the break even = Strike + Option Price) and will thus exercise the option
 - The holder will exercise the option prior to expiry, should the underlying spot exceed the strike of $50. If the spot is less than the break even, exercising the option doesn't yeild a profit but reduces the $2.50 loss.
 - Diagram:

```python
# Import methods
import numpy as np 
import matplotlib.pyplot as plt 

# Define parameters
x_min, x_max = 40, 60
N = 101
S_t = np.linspace(x_min, x_max, N) # Underlying spot at expiry
K = 50 
tc = 2.50 
Payoff = np.maximum(0, S_t - K) - tc 

# Plot 
plt.plot(S_t, Payoff, 'k--', label='Payoff') 
plt.plot([x_min, x_max],[0, 0],'k-',label='$S_{T}$')
plt.xlim([x_min, x_max])
plt.xlabel('Stock Price, S')
plt.ylabel('Payoff')
plt.legend()
plt.show()
```

**Q1.10** - Suppose that a June put option to sell a share for $60 costs $4 and is held until June. Under what circumstances will the seller of the option (i.e. the party with the short position) make a profit? Under what circumstances will the option be exercised? Draw a diagram illustrating how the profit from a short position in the option depends on the stock price at maturity of the option.

Answer:
 - If a party is short a put position, they have sold to another party the right to sell the underlying at the agreed strike and date.
 - The short party makes a profit when the option is not exercised, thus pocketing the premium charged. For the option to not be exercised, the underlying should stay above the strike price of the option, as no counterparty would exercise the option and therefore sell their stocks for less than the higher spot. 
 - If the underlying spot price drops below the strike, the option is likely to be exercised. The seller will make a payoff of: Payoff = Option Price - max(0, Strike - Spot)
 - Diagram:

```python
# Calculations
K = 60 
tc = 4
x_min, x_max = 60-10, 60+10
N = 101
S_t = np.linspace(x_min, x_max, N)
Payoff = -np.maximum(0, K - S_t) + tc

# Plot 
plt.plot([x_min, x_max], [0, 0], 'k-', label='Stock')
plt.plot(S_t, Payoff, 'k--', label='Payoff') 
plt.xlim([x_min, x_max]) 
plt.xlabel('Stock Price, S') 
plt.ylabel('Payoff') 
plt.legend() 
plt.show() 
```

**Q1.18** - Describe the profit from the following portfolio: a long forward contract on an asset and a long European put option on the asset with the same maturity as the forward contract and a strike price that is equal to the forward price of the asset at the time the portfolio is set up.

Answer:
 - The party has bought a forward contract requiring them to purchase/exchange the underlying at the agreed maturity and forward price.
 - The party has bought a put option, giving them the right to sell the underlying at the strike price at maturity. 
 - At maturity, the party must exchange on the forward contract, putting them in possession of the underlying, having paid the agreed forward price. They now have the right to sell the underlying at the same price they bought it for.
 - If the asset is less than the strike, they can sell having only lost the option cost. If the asset is worth more than the strike, they can sell for a profit of Spot - Strike - transaction cost.

```python
# Calculations
F_0 = 20 #assumed forward price
tc = 2 #assumed option cost
x_min, x_max, N = 10,30,101
S_t = np.linspace(x_min, x_max, N)
Option_payoff = np.maximum(0, F_0 - S_t) - tc
Forward_payoff = S_t - F_0
Portfolio_payoff = np.maximum(0, Forward_payoff) - tc

# Plot 
plt.plot(S_t, Option_payoff, 'k--', label='Option Payoff')
plt.plot(S_t, Forward_payoff, 'k-.', label='Forward Payoff')
plt.plot(S_t, Portfolio_payoff, 'k-', label='Payoff') 
plt.xlim([x_min, x_max]) 
plt.xlabel('Asset Price') 
plt.ylabel('Payoff')
plt.grid()
plt.legend()
plt.show()
```

**Q1.19** - In the 1980s, Bankers Trust developed index currency option notes (ICONs). These were bonds in which the amount received by the holder at maturity varied with a foreign exchange rate. One example was its trade with the Long Term Credit Bank of Japan. The ICON specified that if the yen/USD exchange rate, $S_{T}$, is greater than 169 yen per dollar at maturity (in 1995), the holder of the bond receives $1,000. If it is less than 169 yen per dollar, the amount received by the holder of the bond is: \
P = 1000 - max[0, 1000(169/S_T - 1)] \
When the exchange rate is beow 84.5, nothing is received by the holder at maturity. Show that this ICON is a combination of a regular bond and two options.


Answer:
 - Bond structure exists between 84.5 and 169 yen/USD.
 - Below 84.5 there is nothing received and no loss. This amount to a long call option where the user will 'buy-in' to the trade/profit when the yen goes above 84.5
 - Above 169, the profit is capped. Capping the profit is the same as selling the underlying at that price and no longer benefitting from its movements. This is a short call position.

```python
# Calculations
x_min, x_max = 60, 200
S_t = np.linspace(x_min, x_max)
Payoff = (S_t>84.5) * (1000 - np.maximum(0, 1000*((169/S_t) - 1)))

# Plot 
plt.plot(S_t, Payoff, 'k-', label="Payoff")
plt.xlim([x_min, x_max])
plt.xlabel('Yen/USD Exchange')
plt.ylabel('Payoff')
plt.grid()
plt.show()
```

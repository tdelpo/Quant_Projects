### General Notes

 - A **derivative** involves two parties agreeing to a future transaction. Its value depends on (or derives from) the values of other underlying variables.
 - **Exchange-Traded Markets** - is a market where individuals and companies trade standardized contracts that have been defined by the exchange. Traditionally these exchanges used an open outcry system where traders physically met and traded on the floors with shouts and hand signals. Electronic systems have replaced this.
    - Examples include: CME - Chicago Mercantile Exchange, and CBOE - Chicago Board Options Exchange. 
 - **Over-the-counter (OTC) Markets** - two parties typically exchanging derivatives through a central counterparty (CPP). A format typically used by banks, large institutions, etc. where parties can ask/bid on derivatives and then record actions publically. 

### Forward Contracts

 - **Forward Contract** - One of the simplest derivaties. It is an agreement to buy or sell an asset at a certain future time for a certain prices. This can be contrasted with a spot contract, which is a agreement to buy/sell as asset almost immediately. A forward contract is typically traded OTC between two financial institutions or between an institution and one of its clients.
 - One party is **_long_**, and agrees to buy the underlying asset on a certain specified future date for a certain specified price. The other party is **_short_**, agreeing to sell on the same date for the same price.
 - **Payoff from Forward Contract**:
    - Assuming a forward contract obligating the purchase of £1 million for $1,223,000 (or $1.223 USD per GBP) in 6 months time.
    - Scenario 1: USD/GBP spot rises to $1.300 USD/GBP. Forward contract is worth: $1,300,000 - $1,223,000 = $77,000. 
    - Scenario 2: USD/GBP spot drops to $1.200 USD/GBP. Forward contract is worth: $1,200,000 - $1,223,000 = -$23,000.
 - **General Payoff Formula**:
    - Long position payoff = $S_{T} - K$, where K is Strike/Delivery price and S is spot price.
    - Short position payoff = $K - S_{T}$

```python
# Import methods
import matplotlib.pyplot as plt
import math 
import numpy as np

# Calculate parameters
S_0 = 1.223
K = np.linspace(S_0-.1, S_0+.1, 101) # Spot @ t
S = 1.223 # Strike
Payoff_long = (K - S) * 1000000
Payoff_short = (S - K) * 1000000

# Plot results
plt.plot(K, Payoff_long, 'r', label='Long Payoff')
plt.plot(K, Payoff_short, 'b', label='Short Payoff')
plt.plot([S_0, S_0], [-100000, 100000], 'k--', label='Strike')
plt.xlim([S_0-.1, S_0+.1]) 
plt.xlabel('$S_{T}$')
plt.ylim([-100000, 100000])
plt.ylabel('Payoff [$USD]')
plt.legend()
plt.box('on')
plt.grid()
plt.axis('tight')
plt.show()
```

## General Notes:

**Factors Affects Option Prices**:
- The current stock price, $S_{0}$ 
- The strike price, $K$ 
- The time to expiration, $T$ 
- The volatility of the stock price, $\sigma$ 
- The risk-free interest rate, $r$ 
- The dividends that are expected to be paid 


**Stock Price and Strike Price**:
- The difference between the stock price and strike price when exercised, drives the value of an option.
- The more likely that the option expires 'in the money' makes the option more expensive.
- Puts and Calls have opposite relationships

**Time to Expiration**:
- The value of an option can be split into intrinsic and extrinsic value
- The intrinsic value is a measurement of its value if sold today
- The extrinsic value is its potential to gain/lose value with the time left to expiration.

**Volatility**: 
- Determines the degree in which the stock price is expected to potentially move within the given time to expiration
- A more volatile stock can more greater amounts than a less volatily 
- This movement factors into how probably it is that the price will move favourably in reference to the position and strike of the option

**Risk-free Rate**:
- Affects the expected growth of the market in general, and drift of the underlying value
- Increased rates generally increase the expected return of investors and discount future cash flows further. This makes calls more valueable and puts less value (i.e. the movement is expected to favour calls and disadvantage puts)

**Dividends**:
- Affects the future stock price when given.
- Can negatively impact the value of calls while favouring put value

**Option Prices**:
- Similar to futures and forwards, Options have upper and lower bound prices basic on arbitrage arguments
- These bounds can be used to facilitate pricing and recognise mispricings

**Put-Call Parity**:
- Owing to arbitrage opportunities closing, the price of a put and a call are linked and can be used to derive each other.

## Questions:

**Q3.01** - Explain what is meant by a *perfect hedge*. Does a perfect hedge always lead to a beter outcome than an imperfect hedge? Explain your answer.

Answer:
- A perfect hedge eliminates the risk of future downturn by surrendering the benefits of future upturn.
- An imperfect hedge is unable to hedge away the risk and thus keeps some exposure. This means that if the market improves, technically speaking the imperfect hedge can marginally benefit from the movement and out-perform the perfect hedge. 
- Textbook: Think certainty of outcome versus some uncertainty.

**Q3.02** - Under what circumstances does a minimum variance hedge portfolio lead to no hedging at all?

Answer:
- If beta (h*) is equal to zero, there is no slope in the regression between the portfolio and the futures/index. This is pure alpha and the variance is already minimised, thus no further hedge is possible. 
- Think beta = rho * sigma_X/sigma_Y. Variance in x and y are expected, so it is rho that must be zero for beta to be zero. Rho is the coefficient of correlation between the terms, which if zero, means they are orthogonal and entirely independent of each other.

**Q3.06** - Does a perfect hedge always succeed in locking in the current spot price of an asset for a future transaction? Explain your answer.

Answer:
- There will be a difference between the current market spot price and the futures price of the contract. The different is the forward value of the contract given the current price and accounting for market expectation. 
- Furthermore, the value of the futures contract can evolve in time, alongside the evolution of the spot price. This relative movement is the basis risk gap.
- Since basis risk (b2) can't be eliminated/known, the perfect hedge isn't possible.

**Q3.07** - Explain why a short hedger's position improves when the basis strengthens unexpectedly and worsens when the basis weakens unexpectedly.

Answer:
- A short hedger is someone expected to sell an asset in the future who takes a futures position to sell in order to hedge the risk of spot movement.
- When the basis strengthens, this means the value of the futures contracts increases relative to the value of the spot position. This increases the value of the position that is held by the short hedger as they can close their futures contracts for more money than was previously thought.
- When the basis weakens, this means the value of the futures contracts decreases relative to the value of the spot position (e.g. the gap narrows). This decreases the value that the futures position will close for and thus weakens the overall hedge.

**Q3.08** - Imagine you are the treasurer of a Japanese company exporting electronic equipment to the United States. Discuss how you would design a foreign exchange hedging strategy and the arguments you would use to sell the strategy to your fellow executives.

Answer:
- As a company exporting, I am selling assets. The assets are likely paid for in US dollar, and I want to convert them to Yen to balance my booked and note my revenue/profit. 
- When I receive US dollars for my product, I gain exposure to the USD/Yen exchange rate. Assuming the price of my goods don't change, if the USD gets stronger, I will benefit by converting it into more Yen than previous. If it gets weaker, I get less Yen. 
- Hedging this transaction means trying to lock in the price of the USD/Yen exchange rate so that I don't lose when the USD gets weaker. 
- I am expecting to receive USD and sell them. This is a short hedge position. I should sell USD/Yen futures contracts to hedge. 
- To convince the board I'd be interested in whether others in my industry also hedge and to get their buy-in on the hedge so that they don't view it negatively when the company loses exposure to potential upsides.
- Futures/Forwards are used to eliminate downside/upside movement in exchange for more certainty, while Options are used to eliminate downside in exchange for a premium.

**Q3.10** - "If the minimum variance hedge ratio is calculated as 1.0, the hedge must be perfect". Is this statement true? Explain your answer.

Answer:
- A hedge ratio of 1.0, corresponds to a beta of 1.0. This is typical when the asset being traded is the same as the underlying asset on the futures contracts. This occurs because a price movement in spot should near directly impact the futures price of said asset and thus the delta_S and delta_F correlation gives a slope (beta) of 1.
- However, the hedge isn't necessarily perfect. There is still variance, its just been minimised. The error term in the regression highlights this and basis risk still exists.

**Q3.11** - "If there is no basis risk, the minimum variance hedge ratio is always 1.0" Is this statement true? Explain your answer.

Answer:
- True. If there is no basis risk than b2 is know. Since S2 is unknown (e.g the future spot price) it requries a h of 1.0 to eliminate it as a term.

**Q3.12** - "When the futures price of an asset is less than the spot price, long hedges are likely to be particularly attractive" Explain this statement.

Answer:
- A long hedge involves taking a long position in a futures contract. If the futures price is less than the spot price, this allows someone to lock in a cheaper price on the asset in the future rather than buying it more expensively at spot right now. It also defers delivery and storage costs.

**Q3.13** - The standard deviation of monthly changes in the spot price of live cattle is 1.2. The std of monthly changes in futures price of live cattle for the closest contract is 1.4. The producer wants to use December live cattle futures contracts to hedge its risk. Each contract is for the delivery of 40,000 pounds. What strategy should the beef producer follow?

Answer:

**Q3.14** - A corn farmer argues "I do not use futures contracts for hedging. My real risk is not the price of corn. It is that my whole crop gets wiped out by the weather.". Discuss this viewpoint. Should the farmer estimate his or her expected production of corn and hedge to try to lock in a price for expected production?

Answer:
- It may be true that the farmer's biggest and most damaging risk is the weather wiping out his crop. However, that doesn't mean that the price risk isn't real. So let's talk about what might happen if he hedges with futures and then the weather wipes out the crop or not.
- It would be a short hedge to lock in the sale price. Regardless of the sale of the actual asset, the farmer would make a gain if the spot dropped and a loss if the spot increased when he closes out his futures contract. Losing your crop and making a futures loss would hurt the most. But you could also lose your crop and still make a gain on the futures contract. If you don't lose your crop, you've secured your price. Arguably that's 3 good scenarios to one bad. 
- Perhaps a better hedging approach would be a stack and roll? This means the farmer progressively takes positions while he know he has crop, instead of having to pick a larger time period.
- Alternatively, the farmer might prefer options instead of futures/forwards. Going long on a call option at spot would give the farmer certainty that he had product to sell.
- Note that bad weather is likely to affect other farmers too. So the supply will drop and the price/value will increase due to demand remaining the same.

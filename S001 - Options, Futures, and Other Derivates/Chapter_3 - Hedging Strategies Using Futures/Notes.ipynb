## General Notes:

**Basic Principles**:
- The objective of hedging is to reduce the risk of a given position (by as far as possible)
- A **Short Hedge** involves a short position in a futures contract.
    - It is appropriate when the hedger already owns the asset and expects to sell at a time in the future
    - It can also be used when an asset is not currently owned but is expected to be at or before the point of sale
    - The hedge has the affect of locking in the future price of the asset being sold. This protects downside but also eliminates upside.
- A **Long Hedge** involves a long position in a futures contract.
    - It is appropriate when a company expects to purchase an asset in the future and wants to lock the price in now. 
    - As the same for short position, the long futures positions can be closed out in the delivery month.

**Agruments Against Hedging**:
- **Shareholders**: Some argue that shareholders should/could bear the responsibility to act in a way that protects them from their own risks. I.e., that its not the company's responsibility. However, this assumes that the shareholders have all the same information as the company and can then act according to their risk appetite. However, this is rarely possible, and it should also be noted that its rarely financially efficient given individual transaction costs.
- **Competitors**: If it is not the norm within a given industry, hedging may not necessarily make sense for a company. This is because fluctuations in the cost of goods will lead to fluctuations in the price of goods, and, accordingly, if one party is hedging while the rest aren't, they can end up in a worse position.
- **Company Dynamics**: A financial officer will be praised if they hedge risk successfully but might be critisized if they don't (or if was unnecessary). Therefore, the environment of the company and its leadership needs to be aligned.

**Basis Risk**:
- The relative difference in risk position caused by fluctuations in market/future prices and execution timing (e.g. futures expiry).
- Risks that give rise to basis risk include:
    - The asset being hedged may not be exactly the same as the asset underlying the futures contract
    - There may be uncertainty as to the exact date when an asset will be bought or sold
    - The hedge may require a futures contact to be closed out before its delivery month
- **The Basis**:
    - Basis = Spot price of asset to be hedged - Futures price of contract used
    - If asset hedged and asset underlying on the futures contract are the same, the basis should be zero at expiration
    - Can be positive or negative prior to expiration
    - As time evolves, the prices of the two don't necessarily more the same amount, this results in change of basis
    - "Strengthening of the basis" is an increase, and "Weakening of the basis" is a decrease. 

**Basis Equations**:
- Given:
    - $t_{1}$: Enter trade time
    - $t_{2}$: Close trade time
    - $S_{1}$: Spot price at $t_{1}$ 
    - $S_{2}$: Spot price at $t_{2}$ 
    - $F_{1}$: Futures price at $t_{1}$ 
    - $F_{2}$: Futures price at $t_{2}$ 
    - $b_{1}$: Basis at $t_{1}$ 
    - $b_{2}$: Basis at $t_{2}$ 
- By definition:
    - $b_{1}$ = $S_{1}$ - $F_{1}$
    - $b_{2}$ = $S_{2}$ - $F_{2}$
- For example, assume a hedger known that the asset will be sold at t2 and takes a short futures position at t1. The price realized for the asset is S2 and the profit on the futures position is F1 - F2. The effective price via hedging is thus:
    - $S_{2}$ + $F_{1}$ - $F_{2}$ = $F_{1}$ + $b_{2}$ 
    - $b_{2}$ is the unknown and thus basis risk. If $b_{2}$ is known, a perfect hedge would result.
- Again, company knows it will buy the asset at time t2 and initiates a long hedge at time t1. The price paid for the asset is S2 and the loss on the hedge is F1 - F2. This gives the same equation above!
- **Cross Hedging** changes the above equation due to differences in asset being bought/sold versus the one udnerlying the futures contract. The above equation changes:
    - $F_{1}$ + ($S^{*}_{2}$ - $F_{2}$) + ($S_{2}$ - $S^{*}_{2}$) = $F_{1}$ + $b_{2}$
    - Here, ($S^{*}_{2}$ - $F_{2}$) is the basis that would exist if the asset being hedged were the same as the underlying
    - And, ($S_{2}$ - $S^{*}_{2}$) is the basis arising from the difference between the two assets. 

**Choice of Contract**:
- A key factor affecting basis risk is the futures contract chosen. It has two key components:
    - The asset underlying the contract
    - The delivery month
- In practise, the asset can be different to what is being hedged and thus an equivalent asset position and the relationship between the assets needs to be established in order to hedge correctly
- In practise, the futures contract should expire in the months after the underlying is received/sold to avoid the risk of being assigned and taking delivery. This means that the futures contract may not have reached price equalibrium with the underlying but is closed out regardless. This introduces further risk - espeically if price convergence between spot and futures for a given asset is expected much closer to delivery (i.e. the gap is potentially wider).

**Cross Hedging**:
- Hedging an asset with a futures contract whose underlying is a different asset. E.g. An airline hedging jet fuel must use a different fuel/oil since futures in jet fuel aren't actively traded. 
- **Hedging Ratio** is the size of position in futures taken relative to the exposure. 
    - When underlying asset matches hedged, it is normal to use hedging ratio of 1.0. (E.g. Hedge asset is 20,000 barrel of oil and futures position is 20,000 barrels of oil.)
    - When cross hedging, the hedger should choose a value for hedge ratio that minimises variance of the value of the hedge position
    


**Minimum Variance Hedge Ratio**:
- Define:
    - $\Delta S$: Change in spot price, S, during a period of time equal to the life of the hedge
    - $\Delta F$: Change in futures price, F, during a period of time equal to the life of the hedge
- Assuming roughly linear relationship we get:
    - $\Delta S = a + b\Delta F + \epsilon$
    - Where, a and b are constants and $\epsilon$ is an error term (linear regression line equation) 
- Given a hedging ratio, h, Then the chance in value of the position per unit exposure to S is:
    - $\Delta S - h\Delta F = a + (b - h)\Delta F + \epsilon$
    - The standard deviation of this function is minimised by setting b = h. This cancels second RHS term.
- Setting h* to be minimum hedge ratio and having shown that h* = b (minimum variance = minimum deviation), it follows that:
    - $h^{*} = \rho_{SF}\frac{\sigma_{S}}{\sigma_{F}}$
    - Product of coefficient of correlation by ratio of standard deviations. 
    - **Note**: This is the $\beta$ coefficient which can also be expressed as:
        - $\beta = Cov(x,y)/Var(x)$
        - Where, $Cov(x,y) = \rho_{xy}\sigma_{x}\sigma_{y}$
- If $\rho$ is 1.0 and $\sigma_{S}=\sigma_{F}$, then h* is 1.0. This is expected.
- **Hedge effectiveness** is defined as the proportion of variance that is eliminated by hedging. It is the $R^{2}$ of the regression and equals $\rho^{2}$
- The variables are typically calculated from historical data of $\Delta S$ and $\Delta F$. Ideally the data is of the same non-overlapping time periods but this restricts data points, so often small time periods are used.

**Optimal Number of Contracts**:
- Given:
    - $Q_{A}$ = Size of position being hedged (units) 
    - $Q_{F}$ = Size of one futures contract (units) 
    - $N^{*}$ = Optimal number of futures contracts to hedge
- We can calculate the number of contracts required by:
    - $N^{*}$ = $h^{*}\frac{Q_{A}}{Q_{F}}$
    - Round to the nearest integer

**Impact of Daily Settlement**:
- Previous calcualtions have been treated as forward contracts but futures contracts have daily settlement, so minimum variance should take into account this behaviour.
- Given:
    - $\hat{\sigma}_{S}$ = Standard deviation of percentage one-day changes in spot prices
    - $\hat{\sigma}_{F}$ = Standard deviation of percentage one-day changes in futures prices
    - $\hat{\rho}_{SF}$ = Correlation between percentage one-day changes in spot and future prices
    - Note: the percentage change can be calculated with simple or log, but must be consistent between terms
- Optimum hedging ratio and number of contracts is given by:
    - $h^{*}$ = $\hat{\rho}_{SF}\frac{\hat{\sigma}_{S}S}{\hat{\sigma}_{F}F}$
    - $N^{*}$ = $\hat{\rho}_{SF}\frac{\hat{\sigma}_{S}SQ_{A}}{\hat{\sigma}_{F}FQ_{F}}$
    - Note: $SQ_{A}$ = $V_{A}$ is the value of the position (same for futures denotation)
    - To take account of interest earned or paid over life of the hedge, $N^{*}$ and be divided/multipled (e.g. 5%pa would be 1.05). This is referred to as *tailing the hedge*

**Stock Index Futures**:
- Stock indexes can be used to hedge or manage exposures to equity prices
- A **Stock index** tracks changes in a hypothetical portfolio of stocks. The weight of a stock in the portfolio is equal to the proportion of the portfolio invested in that stock. Dividends are not usually included in calculations so that the index tracks the capital gains/loss from investing in the portfolio.
- The hypothetical portfolio of stocks remains fixed, the weights assigned to stocks in the portfolio do not. When one stock increases in relative value more than another, it grains more weight. Sometimes a portfolio is constructed with a set number of shares of each stock, in which can the weights are the values of each individual position. Weighting by market capitalization is also an option. 
- Common indices include: 
    - DOW Jones Industry Average - 30 bluechip US stocks
    - S&P 500 - 500 different companies - 400 industrials, 40 utilities, 20 transportation, and 40 financial.
    - NASDAQ 100 - 100 stocks on the NASDAQ exchange
    - CSI 300 - Chinese stocks


**Hedging an Equity Portfolio**:
- Stock index futures can be used to hedge a well-diversified equity portfolio.
- Given:
    - $V_{A}$ = Current value of the portfolio
    - $V_{F}$ = Current value of one futures contract (futures prices times contract size) 
- If the portfolio mirrors the index, the optimum hedge can be assed to be 1.0 and the number of contracts to be hedged is:
    - $H^{*}$ = $V_{A}$/$V_{F}$
- When the portfolio doesn't mirror the index, we can use the CAPM approach. The parameter beta ($\beta$) from CAPM is the slope of the best-fit line obtained when excess return on the portfolio over the risk-free rate is regressed against the excess return of the index over the risk-free rate. 
    - $\beta$ = 1.0, the portfolio tends to mirror the return on the index
    - $\beta$ = 2.0, The excess return on the portfolio is twice as great as excess return on index
- When a portfolio's $\beta$ is 2.0, it is twice as sensitive to movements in the index and thus needs twice as many contracts to hedge:
    - $H^{*}$ = $\beta\frac{V_{A}}{V_{F}}$
    - This assumes the futures maturity is close to maturity of hedge.
    - Note that comparing equation, $\hat{h}$ = $\beta$

**Reasons for Hedging an Equity Portfolio**:
- Where there is strong correlation between the index and portfolio, hedging the portfolio is likely to result in it growing at merely the risk-free rate. This is questionably advantageous since the investor could just sell the portfolio and invest in risk-free securities for the same return.
- A possible advantage is where the investor is confident in their portfolio but uncertain about the performance of the market - generally feeling that the portfolio will outperform the market. Hedging in this instance removes the risk of the market movement and leaves the investor only exposed to the relative difference in performance between the two. 
- Similarily, a hedge can be used to protect during short periods of uncertainty on a portfolio thats being held for a longer period of time. 

**Changing the Beta of a Portfolio**:
- Futures can be used to change the beta of a portfolio, making it more/less sensitive to index movements
- Given the optimum number of contracts to a minimum variance fit, the investor can take positions in more or less than the optimum to effectively reintroduce variance and index exposure to their portfolio/hedge. 

**Locking in Benefits of Stock Picking**:
- Hedging can be used to lock in benefits of well picked portfolio of stocks. 
- Where it is believe the portfolio will outperform the market in either downturn or upward movement, hedging can yeild gains either way for the investor. 

**Stack and Roll**:
- Stack and roll is the act of choosing futures contracts with earlier expiry than the maturity of a hedge (usually due to preferred liquidity). When a contract matures, it is closed and a new contract opened. Hence, naming. 
- Unlikely to result in a perfect hedge but will reduce some loss incurred by adverse market movements.

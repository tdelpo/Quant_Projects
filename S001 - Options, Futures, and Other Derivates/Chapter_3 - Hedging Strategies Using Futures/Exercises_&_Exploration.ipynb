**Example 3.3**:
- Minimum variance hedging ratio. Keen to explore exactly how the calculations work and to reproduce them.

```python
import pandas as pd
import statistics
```

```python
df = pd.read_excel("Table3_2.xlsx", index_col='Month') 
df.head() 
```

```python
# Calculations
dF_std = statistics.stdev(df.delta_F) 
dS_std = statistics.stdev(df.delta_S) 
rho = statistics.correlation(df.delta_F,df.delta_S)
h_star = rho*dS_std/dF_std
print(dF_std, dS_std, rho, h_star) 
```

## General Notes:

**Types of Options**:
- A **Call** options gives the holder of the option the right to buy an asset by a certain date for a certain price.
- A **Put** option gives the holder the right to sell an asset by a certain date for a certain rpice.
- The date specified is known as the expiration date or the maturity date
- The price specified in the contract is known as the exercise price or the strike price
- American options can be exercised at any time up to the expiry
- European optiosn can only be exercised at expiry

**Properties of Options**:
- Every option has two parties, the short (writer) and the long (buyer/holder) party.
- Long and Short positions have mirrored payoff profiles
- Options can be held on many assets. Common underlyings are Stock, ETP, Currency, Index and Futures.
- Standard terms for options are detailed by the various exchanges on which they sell. These can also be made flexible by the same exchanges.
- There are position and exercise limits to protect against a single party having too much influence
- Market Makers typically facilitate the trade of options providing liquidity by quoting bid/ask prices, agnostic to whether a party is looking to long or short. The difference in the bid/ask price is known as the bid/ask spread and is how the MM makes their money.
- To close an option position, the holder or writer need only short/long the opposite of thier current position. See previous chapter notes about 'open interest' count on positions.
- Options are typically purchased for a fixed fee plus a per-contract fee. There may be fees for exercising options and assigning options. 
- Margin isn't required when buying an option as it doesn't give rise to an obligation (like a foward or futures contract). Margin is required for writing options however.
- **Naked Options** are options sold without owning the underlying asset. This holds more risk.

**Taxation**:
- General rule for retail traders is that gains and losses are treated as capital gains
- Gain or loss is realised when the option is exercised or expires without exercise 
- If exercised, the gain/loss is rolled into the position taken and recognised when the position in the stock/underlying is closed out.
- **Wash Sale Rule** prevents investors from selling and repurchasing stock within a certain window in order to write off losses in their taxes. Using options to work this mechanism is included in the rule.

**Other Notes**:
- **Warrants** are options issued by a financial institution or nonfinancial corporation. They can be issued by a company and attached to other products given to investors.
- **Employee Stock Options** are call options issued to employees by their company to motivate them to act in the best interests of the company's shareholders.
- **Convertible bonds** often referred to as convertibles are bonds issued by a company that can be converted into equity at certin times using predetetermined exchange ratio. They are therefore bonds with embedded call options on the company's stock.

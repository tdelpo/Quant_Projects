## General Notes:

**Notes**:
- An Investment asset is an asset that is held soly for investment purposes by at least some traders.
- A Consumption asset is an asset that is held primarily for consumption. 
- Short Selling is the act of selling an asset that is not owned, typically with the intention of buying it back at a lower bring for a profit. 

**Forward Price of Investment Asset**:
- The forward price of an asset that provides no income:
    - $F_{0} = S_{0}e^{rT}$
    - Where, $F_{0}$ is the forward price, $S_{0}$ is the present spot price, $r$ is the risk-free rate and $T$ is time
- Arbitrage opportunties are realised when the forward price of an asset is out of line with its spot price. 
    - $F_{0} \le S_{0}e^{rT}$: Sell asset and invest proceeds, enter into forward/future contract to buy back at T.
    - $F_{0} \ge S_{0}e^{rT}$: Sell forward/future contract at T, borrow at risk-free to purchase asset at spot right now.


**Forward Price of Investment Asset w/ Known Income**:
- Formula:
    - $F_{0} = (S_{0} - I)e^{rt}$, where $I$ is present discounted value of future known income
- The same arbitrage opportunies present when misalignment of forward price occurs. 


**Forward Price of Investment Asset w/ Known Yield**:
- Formula:
    - $F_{0} = S_{0}e^{(r-q)T}


**Valuing Forward Contracts**:
- Formula:
    - $f = (F_{0} - K)e^{-rT}$
- Used to evaluate the value of a forward contract that at some time $t$ after first entering and before $T$
- Adjusted for each type of forward valuation approach:
    - Asset with no income:
        - $f = S_{0} - Ke^{-rT}$
    - Asset with income:
        - $f = S_{0} - I - Ke^{-rT}$
    - Asset with yeild:
        - $f = S_{0}e^{-qT} - Ke^{-rT}$

**Index Arbitrage**:
- Taking advantage of a misalignment in forward price and spot by buying or selling underlying stocks and buying/selling futures.
- Typically reserved for program trading due to the speed and access required by the traders


**Currency Forward Contract Pricing**:
- Follows same formula principles as others:
    - $F_{0} = S_{0}e^{(r - r_{f})T}$
    - $r$ is domestic risk-free rate and $r_{f}$ is foreign risk-free rate

**Futures on Commodities**:
- Storage costs on commodities act like negative incomes.
- Formula:
    - $F_{0} = (S_{0} + U)e^{rT}$, where $U$ are discounted storage costs
    - If the storage costs are a known proportion to the price of commodity, it can be reformulated as a yield instead.
- Can have convenience yeild, representing value of holding the asset instead of holding futures contracts:
    - $F_{0} = S_{0}e^{(r + u - y)T}$, where $u$ is proportional storage costs and $y$ is convenience yield

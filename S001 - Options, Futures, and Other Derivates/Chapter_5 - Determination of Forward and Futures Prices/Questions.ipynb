## Questions

**Q5.01** - What is the difference between the forward price and the value of a forward contract?

Answer:
- The forward price is the agreed settlement/value of the forward contract for exchanging the prescribed goods
- The value of the forward contract is it's inherent value due to changing in the underlying value of the prescribed goods. If the goods being bought become worth more than agreed in the forward contract, the contract itself gains value since it gets to purchase those goods for less and potentially sell for a profit at spot. The value of the contract decreases if the underlying value move in a negative direction.

**Q5.04** - Explain carefully why the futures price of gold can be calculated from its spot price and other observable variables whereas the futures price of copper cannot.

Answer:
- Gold is an investment asset. If the futures price is too high, investors will find it profitable to increase their holdings of gold and short futures contracts. If it is too low, investors will find it profitable to decrease their holdings and go long on futures contracts. 
- Copper is a consumption asset. If the futures price is too high, the strategy of buy copper and sell futures contracts works. However, because investors do not in general hold the asset, the strategy of sell copper and buy futures is not available to them. There is therefore an upper bound, but no lower bound, to the futures price. 

**Q5.05** - Explain why a foreign currency can be treated as an asset providing a known yield.

Answer:
- A foreign curreny, belonging to a regulated first-world bank, is controlled and subject to a risk-free rate realised in that currency between its national and participating banks. This provides a known yeild i.e. expected growth rate.
- As a currency it can be held by investors as an asset.

**Q5.06** - Is the futures price of a stock index greater than or less than the expected future value of the index? Explain your answer

Answer:
- The futures price of a stock index is always less than the expected future value of the index.
- This is because the expected return on the index is greater than the risk-free rate. Thus:
    - E(St) = S0*e^(u-q)T
    - F0 = S0*e^(r-q)T
    - Since, u > r, E(St) > F0 
    - This is positive systematic risk

**5.14** - Suppose that F1 and F2 are two futures contracts on the same commodity with times to maturity, t1 and t2, where t2 > t1. Prove that F2 <= F1e^R(t2-t1), where r is the interest rate (assumed constant) and there are no storage costs. For the purposes of this problem assume that a futures contract is the same as a forward contract.

Answer:
- F2 = S0e^Rt2, F1 = S0e^Rt1.
- S0 * e^Rt2 = S0 * e^Rt1 * e^R(t2-t1)
- e^Rt2 = e^Rt1 * e^R(t2-t1)
- e^Rt2 = e^(Rt1 + R(t2-t1)) = e^R(t1 + t2 - t1) = e^R(t2)
- If F2 > F1e^r(t2-t1) an arbitrage is created, allowing investor to long t1 futures contract and short t2 futures contract. Profit is F2 - F1e^r(t2-t1). 
- This type of arbirate opportunity cannot exist for long, hence F2 <= F1e^r(t2-t1)

**5.15** When a known future cash outflow i a foreign currency is hedged by a company using a forward contract, there is no foreign exchange risk. When it is hedged using futures contracts, the daily settlement process does leave the company exposed to some risk. Explain the nature of this risk. In particular, consider whether the company is better off using a futures contract or a forward contract when: (a) The value of the foreign currency falls rapidly during the life of the contract. (b) The value of the foreign currency rises rapidly during the life of the contract. (c) The value of the foreign currency first rises and then falls back to its initial value. and (d) The value of the foreign currency first falls and then rises back to it inital value. Assume that the forward price equals the futures price.

Answer:
- Forward contracts are settled at maturity, while Futures are settled daily.
- Assuming I'm operating in native currency, a foreign currency outflow requires me to buy it and sell it. Therefore a fall in the value of the foreign currency increases my purchasing power of it. 
- (a) A loss on the hedge will be realised. I could have bought spot at maturity for less than agreed in my hedge. The forward contract loss is realised at maturity, and the future contract loss is realised daily. The forward contract will be slightly better based on discounted cash value. 
- (b) A gain on the hedge will be realised. Selling at spot at maturity would cost me more than my hedge, thus my hedge has value. I want to realise that value favourably, which is everyday - thus futures is stronger than forward
- (c) Futures is better since it realises positive cash flow early and negative later. 
- (d) Forward is better since it realises all loss at the end.

**Q5.16** - It is sometimes argued that a forward exchange rate is an unbiased predictor of future exchange rates. Under what circumstances is this so?

Answer:
- The forward exchange rate is unbiased predictor of future exchange rate when the exchange rate has no systematic risk. To achieve no systematic risk, the exchange rate must be uncorrelationed (orthogonal) with the return on the market.

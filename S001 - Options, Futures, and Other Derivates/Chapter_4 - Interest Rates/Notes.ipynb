## General Notes:

**Notes**:
- An interest rate defines the amount of money a borrower promises to pay the lender.
- One important factor that impact interest rate is **credit risk** 
- The higher the risk that the money won't be repaid, the higher the credit risk and applied interest rate
- **Credit Spread** is the amount of interest in addition to the risk-free rate
- Interest rates are often expressed in basis points. One pasis point is 0.01% per annum

**Types of Rates**:
- **Treasury Rates**:
    - Rates that investors earn on Treasury bills and bonds. These are instruments used by governments to borrow its own currency. 
    - It is usually assumed that no 1st world developed country will default on a debt obligation in its own currency - this leads to the rates being regarded as "risk-free".
- **Overnight Rates**:
    - Banks are required to maintain a certain amount of cash (reserve) with the central bank. The reserve requirement is dependent on outstanding asset/liabilities at any given time. At the end of the day, this means some institutions will have surplus with the reserve, while others will be in deficit.
    - The overnight rate, or *Federal Funds Rate* is the rate paid by borrowers to lenders overnight in the reserve.
    - Average brokered rates are published - e.g. in the UK, sterling overnight index average (SONIA)
- **Repo Rates**:
    - Unlike overnight fed fund rates, repo rates are secured borrowing rates. 
    - In a repo, an institution that owns securities agrees to sell the securities for a certain price and buy them back later at a slightly higher price.
    - The interest rate is the difference in sale and re-purchase price. 
    - Structured correctly, the repo has reduced credit risk as their is collateral. In theory, the rate is slightly lower than overnight fed rates


**Reference Rates**:
- Typically the agreed rates between two parties of an agreement.
- **LIBOR**:
    - London Interbank Offered Rate - historically a significant reference rate
    - Not enough interbank borrowing meant that rates submitted by contributing parties involved a degree of judgement and could be manipulated. As such, LIBOR being phased out. 
    - LIBOR is a forward looking rate that sets the rate for the period about to occur. New reference rates are backward looking.
    - New reference rates should be considered risk free as they are derived from 1-day loans to creditworthy institutions, while LIBOR had credit spread.

**The Risk-Free Rate**:
- Plays a central role in pricing derivatives as it is argued that a riskless portfolio is setup leading to a risk-free return.
- Typically thought that Treasury bills and bonds are used to derive risk-free rate, however these are artificially low due to favourable tax application.
- The risk-free reference rates created from overnight rates are used in valuing derivatives.

**Measuring Interest Rates**:
- Standard simple interest rate formula:
    - $A(1 + R_{S}/m)^{m*n}$
    - Where, $A$ is invested amount, $R_{S}$ is simple rate of interest, $m$ is compounding periods per year and $n$ is years
- Converting two simple interest rates:
    - $A(1 + R_{1}/m_{1})^{m_{1}*n}$ = $A(1 + R_{2}/m_{2})^{m_{2}*n}$
    - Gives, $R_{2}$ = $4[(1 + R_{1}/m_{1})^{m_{1}/m_{2}} - 1]$
- Compound interest rate formula:
    - $Ae^{R_{c}n}$
    - Where, $R_{c}$ is compound rate of interest
- Converting from simple to continuous:
    - $R_{c}$ = $m \ln(1 + R_{s}/m)$ or,
    - $R_{s}$ = $m(e^{R_{c}/m} - 1)$

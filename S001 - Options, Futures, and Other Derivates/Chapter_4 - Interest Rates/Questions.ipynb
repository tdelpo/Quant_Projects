## Questions:

**Q4.01** - A bank quotes an interest rate of 7% per annum with quarterly compounding. What is the equivalent rate with (a) continuous compounding and (b) annual compounding?

Answer:
- $P$ = A(1 + R/m)^(m*n) = 1*(1 + 0.07/4)^4 = 1.07186
- (a) 1.07186 = Ae^(Rn), R = 0.06939
- (b) 1.07186 = (1 + R), R = 0.07186

**Q4.02** - The 6-month and 1-year zero rates are both 5% per annum. For a bond that has a life of 18 months and pays a coupon of 4% per annum (with semiannual payments and one having just been made), the yield is 5.2% per annum. What is the bond's price? What is the 18-month zero rate? All rates are quoted with semiannual compounding.

Answer:
- Assume face value of $100. 4% per annum coupon is $4, in two $2 payments.
- Bond Price = $2/(1+0.052/2)^(2*0.5) + $2/(1+0.052/2)^(2*1) + $102/(1+0.052/2)^(2*1.5) = 2/1.026 + 2/1.026^2 + 102/1.026^3 = 98.2897
- Solve for 18-month zero rate: 
    - 2/1.025 + 2/1.025^2 + 102/(1+R/2)^3 = 98.2897
    - 102/(1+R/2)^3 = 94.4348
    - 2(sqrt.3(102/94.4348) - 1) = 0.05204

**Q4.15** - Explain carefully why liquidity preference theory is consistent with the observation that the term structure of interest rates tends to be upward-sloping more often than it is downward-sloping.

Answer:
- LPT conjectures that investors favour optionality and liquidity over locking up their funds for extended periods of time. It also rightly points to lenders favouring locking in borrowing for extended period of time. This mismatch in preferences, leads lenders to incentivise borrowers by increasing interest rates on longer term loans. As a result, the term structure of interest rates curves upwards. In terms of financial uncertainty, lenders are less confident in future prospects and thus reduce their incentives to longer commitment periods (they themselves being risk adverse to such commitments), which causes downward sloping term structures.

**Q4.16** - "When the zero curve is upward-sloping, the zero rate for a particular maturity is greater than the par yield for that maturity. When the zero curve is downward-sloping the reverse is true." Explain why this is so.

Answer:
- An upward sloping zero curve assumes growth in the value of money. Par-yield is the interest rate that maintains monetary value. Therefore, upward sloping yield is greater than the par yeild rate, while downward sloping yeild is less than par yeild (as worth has decreasing).
- Book answer: The par yeild is the yield on a coupon-bearing bond. The zero rate is the yield on a zero-coupon bond. When the yield curve is upward sloping, the yield on an y-n-year coupon-bearing bond is less than the yield on an n-year zero-coupon bond. Thus is because the coupons are discounted at a lower rate than the n-year rate and drag the yield down below this rate. Similarily, when the yield curve is downward sloping, the yield on an n-year coupon bearing bond is higher than the yield on an n-year zero-coupon bond becuase the coupons are discounted at relatively higher rates.

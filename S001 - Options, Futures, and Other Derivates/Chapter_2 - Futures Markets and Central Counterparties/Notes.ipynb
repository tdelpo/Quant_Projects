## General Notes:

**Intro Notes**:
 - A vast majority of future contracts do not lead to delivery as they are closed out prior to expiry by purchasing the opposing contract. A profit (or loss) is generated by the spread in the purchase and sale price of the position. 
 - **Long Futures Position** is when a party purchases a futures contract to buy
 - **Short Futures Position** is when a party sells a futures contract to sell
 - Futures contracts requires specifications:
    - **The Asset** - specifies details of the asset that is being traded, including grades and allowances for variations, plus agreed price variations.
    - **Size** - specifies the amount of the asset that is being traded
    - **Delivery Arrangements** - specifies, if any, the delivery arrangements of the asset
    - **Delivery Month** - specifies the period of time in a given month when a delivery can be made
    - **Price Quotes** - specifies how the assets and contracts will be quotes (e.g. what currency and weight) 
 - **Price Limits** - exchanges will restrict the amount by which the price of a futures contract can increase or decrease within a given day. When this occurs, trading holts for the day and the contracts are referred to as "limit up" or "limit down".
 - **Position Limits** - exchanges will restrict the volume of contracts which any one party can hold, so as to not exert undue influence on the market.
 - **Futures Price & Spot Price Convergence** - Closer to the delivery period, the futures price will converge to the spot price. This occurs due to potential arbitrage opportunities prior to convergence.
    - If the futures price is above spot, traders will sell futures, buy spot and make delivery.
    - If the futures price is below spot, traders will buy futures and wait for delivery. 
    - Both these actions encourage the futures price to merge to spot

**Margin Accounts**:
- Traders using an exchange and pruchasing/selling futures are required to maintain a margin account. This is the exchange's way of holding insurance on the buyer/seller's capacity to fullfill their obligation at maturity.
- **Initial Margin** - The initial sum of money/value that the exchange requires the trader to post to their account as collateral.  
- **Daily Settlement** - At the end of each trading day, the account is adjusted to reflect the present value of the contract. This is referred to as "daily settlement" or "marking to market".
- **Variation Margin** - the daily fluctuation of value in/out of the margin account due to marking to market activities. 
- **Maintenance Margin** - An amount below the initial margin, where if the account drops below said value, the exchange will post a "margin call" requiring the trader to topup their account.
- **Cash Interest** - Cash provided to the margin account typically attracts interest payments. Securities can be posted for margin instead, and their value for margin purposes is a fraction of their market value (the reduction is referred to as a "haircut"). 

**Clearing House**:
- A **clearing house** acts as an intermediary in futures transactions.
- Role is to guarantee performance of the parties in transactions.
- Institutions can be members of the clearing house and must be to conduct transactions. Brokers that aren't, are thus required to channel their transactions through someone who is. 
- The clearing house typically calculates posted margin requirements so that they are approximately 99% sure of transactions and losses being covered. 
- The purpose of margin is to be able to pay traders when they make a profit.

**OTC Markets**:
- Are markets where companies agree to derivatives transactions without involving an exchange.
- Direct deals mean that counterparties are exposed to credit risk (default), requiring measures to be taken
- **Central Counterparties (CCP)** - are clearing houses for standard OTC transactions, performing same role as exchange. Once two parties agree a deal, they can present the deal to the CCP. Assuming it is accepted by the CCP, the CCP become the counterparty to both and guarantees the transaction. Therefore, members provide funds to the CCP which contribue to its guarantee fund. 
- Members of a CCP provide initial margins and daily settlements.
- **Bilateral Clearing** - Agreements between two companies which are cleared between the parties themselves. Often involves extended agreement which includes collateral requirements, margins and settlements. 


**Market Quotes**:
- Futures quotes are available from exchanges and several online sources. Details include:
    - Prices - typically showing open, high and low price of trading in the day so far.
    - Prior Settlement Price - price used to calculate daily gains, losses and margin requirements. Typically close of last day.
    - Last trade - price of most recent settled contract of the day
    - Change - difference between last and prior
    - Volume - The number of contracts traded in a day.
- A futures market where prices increase with maturity is known as a "normal market"
- A futures market where prices decrease with maturity is known as an "inverted market" 

**Types of Traders & Types of Orders**:
- Two types of traders executing traders: Futures commissions merchants & Locals. FCMs execute trades on behalf of clients and receive commission. Locals execute trades on their own accounts. 
- **Speculators** can be classed as scalpers, day traders, or position traders.
    - **Scalpers** - monitor very short-term trends and attempt to profit from small changes in the contract price. They hold their positions for only a few minutes.
    - **Day Traders** - hold their positions for less than a day and are generally unwilling to take risk that adverse news will occur overnight. 
    - **Position Traders** - hold positions for longer periods of time. They target profit from major movements in the markets
- **Orders**:
    - **Market Order** - immediately order to carry out trader at the best available price
    - **Limit Order** - the order can only be executed at the given price or more favourable price
    - **Stop Order/Stop-loss Order** - specifies an ask/bid price which activates the order, making the order now a market order. 
    - **Stop-limit Order** - combination of limit and stop. Once stop reached, becomes a limit order. 
    - **Market-if-touched Order** - is executed at best available price after a trade occurs at the specified price or more favourable price.
    - **Discretionary Order** - or "market-not-held order" is traded as a market order except that execution may be delayed by broker in an attempt to get a better price. 
    - **Fill-or-kill Order** must be filled immediately on receipt or not at all
    - Orders expire at the end of a trading day unless specified. 

**Cornering the marker** - An occassional trading irregularity where a trading group takes a large long futures position while simultaneously trying to exercise some control over the supply of the underlying commodity. At the maturity of the futures position approaches, the trader group doesn't close their position, leading to a shortfall/imbalance of short/long positions and a scarcity of supply. As short sellers panic to close their positions (which drives up long futures value), the trading group profits from their long position gaining value. Regulators try to stop this action by imposing size/volume limits and increasing margin requirements on accounts.

**Forward vs Futures Summary**:
- **Forward**:
    - Private contract between two parties
    - Not standardized
    - Usually one specified delivery date
    - Settled at the end of contract
    - Delivery or final cash settlement usually takes places
    - Some credit risk
- **Futures**:
    - Traded on an exchange
    - Standardized contract
    - Range of delivery dates
    - Settled Daily
    - Contract is usually closed out prior to maturity
    - Virtually no credit risk

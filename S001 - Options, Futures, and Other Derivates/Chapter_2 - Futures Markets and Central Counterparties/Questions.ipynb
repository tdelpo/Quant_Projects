**Q2.01** - Suppose that you enter into a short futures contract to sell July silver for $17.20 per ounce. The size of the contract is 5,000 ounces. The initial margin is $4000, and the maintenance margin is $3000. What change in the futures price will lead to a margin call? What happens if you do not meet the margin call?

Answer:
- Silver is $17.20 per ounce. Contract is $5,000 ounces. This gives contract value of $86,000. 
- Initial margin is $4,000. Maintenance at $3000. Margin call occurs when 4000-3000=1000 is lost from the account
- If silver increases by $1000 / 5000 contracts = $0.20 per ounce. Or, $17.40 Silver
- If margin call is not met, the broker closes your account with a market order to buy a futures contract for July at current futures price.

**Q2.02** - Suppose that in September 2021 a company takes a long position in a contract on May 2022 crude oil futures. It closes out its position in March 2022. The futures prices (per bareel) is $48.30 when it enters into the contract, $50.50 when it closes out its position, and $49.10 at the end of December 2021. One contract is for delivery of 1,000 barrels. What is the company's total profit? When is it realized? How is it taxed if it is (a) a hedger and (b) a speculator? Assume that the company has a December 31 year end.

Answer:
- The company buys in September, a May contract, for $48.30 per barrel
- The company sells in March, a May contract, for $50.50 per barrel
- In December, the position is worth $49.10 per barrel
- Profit made is $50.50-48.30 = $2.20 per barrel. 
- Total profit is realised on a day-by-day basis across the two financial years
- A hedger would be taxed in March on the full profit. A speculator would be taxed on 2021 and 2022 profits separately.

**Q2.03** - What does a stop order to sell at $2 mean? When might it be used? What does a limit order to sell at $2 mean? When might it be used?

Answer:
- A stop order to sell at $2 is an order to sell at the best available price once a price of $2 or less is reached
- It could be used to limit the losses of an existing long position - e.g. it detects a market downward trend and sells before it erodes profits.
- A limit order to sell at $2 is an order to sell at a price of $2 or more.
- It could be used to instruct a broker that a short position should be taken, providing it can be done at a price at least as favorable as $2.

**Q2.05** - The party with a short position in a futures contract sometimes has options as to the precise asset that will be delivered, where delivery will take place, when delivery will take place, and so on. Do these options increase or decrease the futures price? Explain your reasoning?

Answer:
- An exchange will typically determine the flexibility which the short position holder has in regard to these characteristics. To the point that they will specify set rates for increased/decreased value received for the position based on the properties of what is delivered (whenand where).
- However, it is expected that the short party will act in their interest and decrease cost to themselves, thus they will likely delivery the most cost effective product to themselves primarily.
- The risk is that this doesn't match the most desired position of the long party. And thus, the long party see potentially less value being realised which it also doesn't have control over. 
- Therefore, I think the variability leads to decreased inherent value in the contract. This decreases the value of the futures price. 

**Q2.06** - Explain how margin accounts protect futures traders against the possibility of default.

Answer:
- Margin account requirements are typically set by the exchange and are expected to de-risk default likelihood to the 99% mark.
- By setting the margin requirements across the exchange. The exchange holds sufficient funds to cover defaulted positions and also pay traders their profits
- Constantly adjusting via margin calls and holding the specified amount makes it unlikely that the chain of parties will default.

Answer:

**Q2.09** - Explain the difference between a market-if-touched order and a stop order

Answer:
- A market-if-touched order is executed at the best available price after a **trade** occurs at the specified price or at a more favourable price.
- A stop order is executed after there is a **bid or offer** at the specific price or at a price **less** favourable than the specified price. 
- Market-if-touched therefore acts when there is confirmed trade volume/activity at a given price or more favourable
- A stop order therefore acts when there is confirmed intension/interest at a given price. 

**Q2.10** Explain what a stop-limit order to sell at 20.30 with a limit of 20.10 means

Answer:
- A stop order activates into a market order when an ask/bid is detected at a prescribed value
- A limit order is executed at a given value or more favourable
- A stop-limit to sell at 20.30 with a limit of 20.10 means that as soon as there's a bid at 20.30, the contract should be sold provided it can be done at or above 20.10

**Q2.01** - Suppose that you enter into a short futures contract to sell July silver for $17.20 per ounce. The size of the contract is 5,000 ounces. The initial margin is $4000, and the maintenance margin is $3000. What change in the futures price will lead to a margin call? What happens if you do not meet the margin call?

Answer:
- Silver is $17.20 per ounce. Contract is $5,000 ounces. This gives contract value of $86,000. 
- Initial margin is $4,000. Maintenance at $3000. Margin call occurs when 4000-3000=1000 is lost from the account
- If silver increases by $1000 / 5000 contracts = $0.20 per ounce. Or, $17.40 Silver
- If margin call is not met, the broker closes your account with a market order to buy a futures contract for July at current futures price.

**Q2.03** - What does a stop order to sell at $2 mean? When might it be used? What does a limit order to sell at $2 mean? When might it be used?

Answer:
- A stop order to sell at $2 is an order to sell at the best available price once a price of $2 or less is reached
- It could be used to limit the losses of an existing long position - e.g. it detects a market downward trend and sells before it erodes profits.
- A limit order to sell at $2 is an order to sell at a price of $2 or more.
- It could be used to instruct a broker that a short position should be taken, providing it can be done at a price at least as favorable as $2.

**Q2.06** - Explain how margin accounts protect futures traders against the possibility of default.

Answer:

**Q2.09** - Explain the difference between a market-if-touched order and a stop order

Answer:
- A market-if-touched order is executed at the best available price after a **trade** occurs at the specified price or at a more favourable price.
- A stop order is executed after there is a **bid or offer** at the specific price or at a price **less** favourable than the specified price. 
- Market-if-touched therefore acts when there is confirmed trade volume/activity at a given price or more favourable
- A stop order therefore acts when there is confirmed intension/interest at a given price.

**Q2.11** - At the end of one day a clearing house member is long 100 contracts, and the settlement price is $50,000 per contract. The original margin is $2,000 per contract. On the following day the member becomes responsible for clearing an additional 20 long contracts, entered into at a price of $51,000 per contract. The settlement price at the end of this day is $50,200. How much does the member have to add to its margin account with the exchange clearing house?

Answer:
- New initial margin = 20 & $2,000 = $40,000
- Gain value of old = ($50,200 - $50,000) * 100 = $20,000 
- Loss value on new = ($50,200 - $51,000) * 20 = $16,000 
- Account requirement = $40,000 - $20,000 + $16,000 = $36,000 

**Q2.12** - Explain why collateral requirements increased in the OTC market as a result of regulations introduced since the 2008 financial crisis.

Answer:
- The OTC markets were originally predominantly direct trades between counterparties not requiring an exchange/clearing and thus margin. Collateral was agreed between the counterparties directly and in-linewith their individual appetites for risk.
- Regulations sort to increase both standardization and counterparty insurance of the OTC market through involving exchange-like clearance houses and centralisation of the contracts being agreed/recorded, and as a result the central counterparties introduced margins in more alignment with exchanges.

**Q2.15** - "Speculation in futures markets is pure gambling. It is not in the public interest to allow speculators to trade on a futures exchange." Discuss this viewpoint.

Answer:
- Speculation is the act of taking positions in the market to benefit from movements in the price, rather than to actually necessarily acquire the underlying asset being traded. 
- The different types of speculation include scalping, day-trading and long/short/position traders. Each of these affect the market in different timescales.
- Speculators, like arbitrageurs, arguably contribute to the efficient market thesis/mechanism by encouraging the price of the underlying to move and fall inline with its 'true' market value and in doing so, having contributed to the information within the market. This is the basis of a fair price, even for consumers of the underlying.
- Note: they add "liquidity" and regulators generally only allow speculating mechanism/products when they're of interest to hedgers as well as speculators. Hedging is de-risking and smart business acument. A product that is purely speculative is vulnerable to a 'gambling' argument.

**Q2.17** - "When a futures contract is tracked on the floor of the exchange, it may be the case that the open interest increases by one, stays the same, or decreases by one.". Explain this statement.

Answer:
- If both parties are entering into a new contract, the open interest increases by one.
- If both parties are exiting/closing a contract/position, the open interest decreases by one.
- If one party is opening while the other is closing, the interest stays the same.

**Q2.19** - A cattle farmer expects to have 120,000 pounds of live cattle to sell in 3 months. The live cattle futures contract traded by the CME Group is for the delivery of 40,000 pounds of cattle. How can the farmer use the contract for hedging? From the farmer's viewpont, what are the pros and cons of hedging?

Answer:
- The farmer can enter into 3 contracts to sell in 3 months time his cattle, instead of selling the cattle on the spot market.
- By entering into a futures contract, the farmer hedges his risk by guaranteeing his settlement price in the 3 months time, rather than risking that the spot market price on that date is less than the current spot/futures price.
- The pros of taking the hedge is derisking exposure to fluctuating spot prices. The cons are that if the spot increases, the farmer can't realise this profit unless it increases enough that he can buy back his futures position, sell on spot and still make additional profit.

**Q2.21** - Explain how CCPs. What are the advantages to the financial system of requireing CCPs to be used for all standardized derivatives transactions between financial institutions.

Answer:
- CCPs stand for Central CounterParties.
- These are a conglomerate of financial institutions/members which participate in the CCP and thus pay membership. The CCP acts as the counterparty between financial instittutions instead of direct deals between them. This provides several benefits:
    - The CCP guarantees against defaulting activities and thus provides stability and insurance.
    - The CCP polices the quality and standardization of the contracts, as well as creating visibility and regulation.
- This absorbs credit risk and means that the market is less likely to take a loss when a major player defaults
- The disadvantage of CCPs are that they are replacing banks as the to-big-to-fail entitities in the financial system.
